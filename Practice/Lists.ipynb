```python
result_list =["A+", "A", "B+", "B", "C+", "C", "D+", "D"]
newlist = []

for x in result_list:
  if "C" in x:
    newlist.append(x)
print(newlist)
```

```python
result_list =["A+", "A", "B+", "B", "C+", "C", "D+", "D"]
newlist = [x for x in result_list if "C" in x]
print(newlist)
```

```python
def myfunc(n):
  return abs(n-50)

Markes = [42, 50, 21,54,74,99,14]
Markes.sort(key = myfunc)
print(Markes)
```
