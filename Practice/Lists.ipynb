```python
result_list =["A+", "A", "B+", "B", "C+", "C", "D+", "D"]
newlist = []

for x in result_list:
  if "C" in x:
    newlist.append(x)
print(newlist)
```

```python
result_list =["A+", "A", "B+", "B", "C+", "C", "D+", "D"]
newlist = [x for x in result_list if "C" in x]
print(newlist)
```
