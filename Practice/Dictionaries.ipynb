```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.keys()
y = car.values()
z = car.items()

print(x)
car["color"] = "white"
car["year"] = 2000
print(x)
print(y)
print(z)
```
